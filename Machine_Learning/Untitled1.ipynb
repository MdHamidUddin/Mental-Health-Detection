```python
import pandas as pd
from imblearn.over_sampling import SMOTE

# for reproducibility purposes
seed = 100
# SMOTE number of neighbors
k = 1

df = pd.read_csv("Dataset3.csv")
df.describe()
```

```python
# make a new df made of all the columns, except the target class
X = df[['Age_of_Subject', 'Time_spent_on_Online_Class', 'Rating_of_Online_Class_experience','Medium_for_online_class', 'Time_spent_on_self_study', 'Time_spent_on_fitness','Time_spent_on_sleep','Time_spent_on_social_media','Number_of_meals_per_day','Change_in_your_weight','Health_issue_during_lockdown','Time_utilized']]
y = df['Time_spent_on_social_media']
sm = SMOTE(sampling_strategy='auto', k_neighbors=k, random_state=seed)
X_res, y_res = sm.fit_resample(X, y)


df = pd.concat([pd.DataFrame(X_res), pd.DataFrame(y_res)], axis=1)

df.to_csv('df_smoted.csv', index=False, encoding='utf-8')
```
